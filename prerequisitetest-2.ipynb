1.List Operations and Basic Statistics

```python
natural_no=[x for x in range(1,21)]
sum=0
for i in natural_no[1::2]:
    sum+=i
print(sum)
product=1
for i in natural_no[::2]:
    product*=i
print(product)
total=(20*(20+1))/2
mean=total/20
print(mean)
median=(natural_no[9]+natural_no[10])/2
print(median)
deviated_squared=0
for i in natural_no:
    deviated_squared+=(i-mean)**2
standard_deviation=(deviated_squared/20)**0.5
print(standard_deviation)
```

2.Mutable vs. Immutable Data Types

```python
'''Mutable Datatypes Vs Immutable Datatypes
Mutable data types are those whose value can be changed or updated without creating a new data type. Example: List; in the List data type, we can update any element without creating a new list with the updated element.
Immutable Datatype are those datatypes whose values can be changed or updated. Example: String; unlike a list, to update a string element, we need to create a whole new string.'''
```

3. Fibonacci Sequence

```python
n=int(input())
fib=[x for x in range(n)]
fib[0]=1
fib[1]=1
for i in range(2,n):
    fib[i]=fib[i-1]+fib[i-2]
print(*fib)
import matplotlib.pyplot as plt
import numpy as np
x=np.array([x for x in range(1,n+1)])
y=np.array(fib)
plt.plot(x,y)
plt.show()
```

7.Dictionary Operations

9.Merging Dictionaries

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
print(dict1,dict2)
for key, value in dict1.items():
    if key in dict2:
        dict2[key]+=value
    else:
        dict2[key]=value
print(dict2)
```

11.Remove Duplicates

```python
elements=[3, 5, 7, 3, 2, 5, 8, 7, 9, 3]
unique_elements=[]
for i in elements:
    if i not in unique_elements:
        unique_elements.append(i)
import numpy as np
print(np.mean(unique_elements))
print(np.median(unique_elements))
```

12.String Operations

```python
data="Hello, welcome to the world of Python!"
data=data.split()
length_counter={}
for i in data:
    length_counter[i]=len(i)
print(max(length_counter,key=length_counter.get))
count=0
for i in data:
    if i[0].lower() in ["a","e","o","u","i"]:
        count+=1
print(count)
import matplotlib.pyplot as plt
x=list(length_counter.keys())
y=list(length_counter.values())
plt.bar(x,y)
plt.show()
```

16.Set Operations:

```python
a=[1,2,3,4,5,6,7,8,9]
b=[2,4,6,8,12,16]
a=set(a)
b=set(b)
intersection=a.intersection(b)
print(intersection)
union=a.union(b)
print(union)
symmetric_diff=a.symmetric_difference(b)
print(symmetric_diff)
from matplotlib_venn import venn2
import matplotlib.pyplot as plt
venn2(subsets=(len(a),len(b),len(intersection)))
plt.show()
```

18.Word Count and Frequency:

```python
with open("text.txt","r") as f:
    lencounter={}
    for i in f.readlines():
        s=(i.strip()).split()
        for i in s:
            lencounter[i]=lencounter.get(i,0)+1
print(max(lencounter,key=lencounter.get))
x=list(lencounter.keys())
y=list(lencounter.values())
import matplotlib.pyplot as plt
plt.bar(x,y,width=0.5)
plt.show()
```

19.Random Number Generation:

```python
import random
data=[random.randint(1,50) for i in range(100)]
with open("random.txt","w") as f:
    for i in data:
        f.write(f"{i}\n")
def strip(x):
    return int(x.strip())
with open("random.txt","r") as f:
    data=list(map(strip,f.readlines()))
print(data)
import matplotlib.pyplot as plt
plt.hist(data,bins=50)
plt.show()
```

22.Normalization:

```python
import matplotlib.pyplot as plt 
data=[12, 45, 23, 89, 56, 34, 78, 91, 67, 42]
minimum=min(data)
maximum=max(data)
diff= maximum-minimum
normalised_data=[]
for i in data:
    normalised_data.append((i-minimum)/diff)
print(normalised_data)
plt.plot(data,normalised_data)
```
