1.List Operations and Basic Statistics

```python
natural_no=[x for x in range(1,21)]#to get a list of first 20 natural nos
sum=0
for i in natural_no[1::2]:#getting odd nos
    sum+=i
print("sum is:",sum)
product=1
for i in natural_no[::2]:#getting even nos
    product*=i
print("product is:",product)
total=(20*(20+1))/2
mean=total/20
print("mean is:",mean)
median=(natural_no[9]+natural_no[10])/2
print("median is",median)
deviated_squared=0
for i in natural_no:
    deviated_squared+=(i-mean)**2
standard_deviation=(deviated_squared/20)**0.5
print("standard deviation is:",standard_deviation)
```

2.Mutable vs. Immutable Data Types

```python
'''Mutable Datatypes Vs Immutable Datatypes
Mutable data types are those whose value can be changed or updated without creating a new data type. Example: List; in the List data type, we can update any element without creating a new list with the updated element.
Immutable Datatype are those datatypes whose values can be changed or updated. Example: String; unlike a list, to update a string element, we need to create a whole new string.'''
```

3. Fibonacci Sequence

```python
n=int(input())#taking input
if n<1:
    raise ValueError("The number of terms must be a positive integer.")
fib=[x for x in range(n)]#making a list of n nos.
fib[0]=1
fib[1]=1
for i in range(2,n):
    fib[i]=fib[i-1]+fib[i-2]
print("Fibonacci series:",*fib)
import matplotlib.pyplot as plt
import numpy as np
x=np.array([x for x in range(1,n+1)])
y=np.array(fib)
plt.plot(x,y)
plt.title("Fibonacci sequence")
plt.xlabel("sequence")
plt.ylabel("Fibonacci value")
plt.show()
```

9.Merging Dictionaries

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
print(dict1,dict2)
for key, value in dict1.items():
    if key in dict2:
        dict2[key]+=value
    else:
        dict2[key]=value
print("Output:",dict2)
```

10.Character Frequency

12.String Operations

```python
data="Hello, welcome to the world of Python!"
data=data.split()
length_counter={}
for i in data:
    length_counter[i]=len(i)#making a dictionary with word :  length 
print("word with max length:",max(length_counter,key=length_counter.get))#finding max length
count=0
for i in data:
    if i[0].lower() in ["a","e","o","u","i"]:
        count+=1
print("words staring with vowels:",count)
import matplotlib.pyplot as plt
y=list(length_counter.values())
plt.hist(y,bins=10)
plt.title("string length distribution")
plt.show()
```

18.Word Count and Frequency:

```python
with open("text.txt","r") as f:
    lencounter={}
    for i in f.readlines():
        s=(i.strip()).split()#removing white sapces and newline characters
        for i in s:
            lencounter[i]=lencounter.get(i,0)+1
print("most frequent word is:",max(lencounter,key=lencounter.get))
x=list(lencounter.keys())
y=list(lencounter.values())
import matplotlib.pyplot as plt
plt.bar(x,y,width=0.5)
plt.title("word frequency visualisation")
plt.xlabel("words")
plt.ylabel("frequency")
plt.show()
```

19.Random Number Generation:

```python
import random
import numpy as np
data=[random.randint(1,50) for i in range(100)]#generating a random set of nos. in range 1 to 50

with open("random.txt","w") as f:
    for i in data:
        f.write(f"{i}\n")
def strip(x):
    return int(x.strip())
with open("random.txt","r") as f:
    data=list(map(strip,f.readlines()))
print("mean of data:",np.mean(data))
import matplotlib.pyplot as plt
plt.hist(data,bins=50)
plt.title("number distribution")
plt.xlabel("number")
plt.ylabel("frequency")
plt.show()
```

22.Normalization:

```python
import matplotlib.pyplot as plt 
data=[12, 45, 23, 89, 56, 34, 78, 91, 67, 42]
minimum=min(data)
maximum=max(data)
diff= maximum-minimum
normalised_data=[]
for i in data:
    normalised_data.append((i-minimum)/diff)#normalising the data in 0 to 1 range using linear normalisation
print("normalised data",normalised_data)
plt.plot(data,normalised_data)
plt.title("normalised data visualisation")
plt.xlabel("given numbers")
plt.ylabel("normalised numbers")
plt.show()
```
